### Group name: Mental Health (10+/d)

To prevent the account from being temporarily suspended, need to:
- Reduce the calls per request as much as possible 
- Add a sleep period after every request

```python
# Instantiate empty DF to store info
df_full1=pd.DataFrame()
```

```python
%%time
#Scraping 10 pages

i=1
for post in get_posts(group='541132720323699', extra_info=False, posts_per_page=200, options={"comments":False, "reactors": False, "allow_extra_requests": False}, pages=10,credentials=('xx@gmail.com','xx!')):
    post_entry = post
    df = pd.DataFrame.from_dict(post_entry, orient='index')
    df = df.transpose()
    df_full1 = df_full1.append(df)
    print(f'Getting post {i}...2 Second sleep')
    i+=1
    time.sleep(2)
```

```python
df_full1['time'].min()
```

**Scraping 10 pages in a group that posts 10 posts a day, returns 210 posts and takes the date back to September 15**
